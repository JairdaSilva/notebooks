```julia
n = 5
src,dst = 1:n, n+1:2n
nms = vcat([" x$i " for i=1:n], [" y$i " for i=1:n])
kw = makekw(
    names=nms,
    x=vcat(zeros(n),ones(n)),
    y=repmat(reverse(linspace(0,1,n)),2),
    lc=:orange,
curves=false,
    arrow=true,shorten=0.04,w=2
)

plotbox()
graphplot!(src, dst; title="relu: y = max(0, x)", kw...)
```

```julia
p = plotbox()
src,dst=Int[],Int[]
for i=1:n, j=1:n
    push!(src,i)
    push!(dst,n+j)
end
graphplot!(src, dst; title="affine: y = wx + b", kw...)

# b
bx,by = .5, -.05
for i=1:n
    plot!([bx,1],[by,(i-1)/(n-1)],l=(2,:orange))
end
scatter!([bx],[by],m=(20,:cyan,.99,:rect),text=["b"])

# w
plot!([.45,.55,.55,.45],[0,0,1,1],t=:shape,ann=(.5,.5,text("w")),c=:cyan)
```

```julia
kw = makekw(ms=(2,1), shorten=0.2, arrow=true,root=:bottom)
```

```julia
g = @flow affine(x) = w * x + b
plot(g; kw..., size=(400,300))
```

```julia
plotbox()
plot!(g, ms=(3,1), shorten=.04, arrow=true, x=[1,.66,.33,0,.5],y=[.5,.6,.9,.5,.1], 
title="affine: y = wx + b", mc=[:orangered,:lightgray,:cyan,:orangered,:cyan], l=(2,:orange))
```

```julia
g = @flow f(x) = out(w11*x1 + w12*x2 + b1, w21*x1 + w22*x2 + b2)
plot(g; kw..., size=(400,300))
```

```julia
plot(@flow f(x) = affine(x); kw..., size=(400,300))
```

```julia
g = @flow net(x) = logistic(w2*relu(w1*x+b1)+b2)
plot(g; kw...)
```

```julia
g = @flow net(x) = logistic(affine(relu(affine(x))))
plot(g; kw...)
```

```julia
g = @flow function net(x)
    hidden = relu( Wxh*x + Whh*hidden + bh )
    y = logistic( Why*hidden + Wxy*x + by )
end
p1 = plot(g; kw..., size=(400,300), title="Tensor Computation Graph")
```

```julia
g = @flow function net(x)
    hidden = relu(affine(concat(x,hidden)))
    y = logistic(affine(hidden))
end
p2 = plot(g; kw..., size=(400,300), title="Transformations")
```
