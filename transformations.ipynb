```julia
n = 5
src,dst = 1:n, n+1:2n
nms = vcat([" x$i " for i=1:n], [" y$i " for i=1:n])
kw = makekw(
    names=nms,
    x=vcat(zeros(n),ones(n)),
    y=repmat(reverse(linspace(0,1,n)),2),
    lc=:orange,
curves=false,
    arrow=true,shorten=0.04,w=2
)

plotbox()
graphplot!(src, dst; title="relu: y = max(0, x)", kw...)
```

```julia
p = plotbox()
src,dst=Int[],Int[]
for i=1:n, j=1:n
    push!(src,i)
    push!(dst,n+j)
end
graphplot!(src, dst; title="affine: y = wx + b", kw...)

# b
bx,by = .5, -.05
for i=1:n
    plot!([bx,1],[by,(i-1)/(n-1)],l=(2,:orange))
end
scatter!([bx],[by],m=(20,:cyan,.99,:rect),text=["b"])

# w
plot!([.45,.55,.55,.45],[0,0,1,1],t=:shape,ann=(.5,.5,text("w")),c=:cyan)
```

```julia
g = @flow function net(x)
    hidden = logistic( Wxh*x + Whh*hidden + bh )
    y = logistic( Why*hidden + Wxy*x + by )
end
plot(g; kw...)
```

```julia
g = @flow function net(x)
    hidden = logistic(affine(concat(x,hidden)))
    y = logistic(affine(hidden))
end
plot(g; kw...)
```
